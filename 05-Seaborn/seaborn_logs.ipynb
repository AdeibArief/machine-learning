# Categorical plots

```python
df=pd.read_csv("./StudentsPerformance.csv")
df
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,x='reading score',y='parental level of education',hue='test preparation course')
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(data=df,x='reading score',y='parental level of education',inner='quartile',bw=0.5)
```

```python
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde',hue='gender',shade='True')
```

```python
sns.pairplot(data=df,hue='gender')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='violin',col='lunch',row='race/ethnicity')
```

```python
g=sns.PairGrid(df,hue='gender')
g.map_upper(sns.kdeplot)
g.map_diag(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.add_legend()
```
