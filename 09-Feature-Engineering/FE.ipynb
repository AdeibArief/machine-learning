```python
def sample_ages(mu=50,sigma=13,num_samples=100,seed=42):
        np.random.seed(seed)
        ages = np.random.normal(loc=mu,scale=sigma,size=num_samples)
        ages= np.round(ages,decimals=0)
        return ages
```

### Dealing with missing data

```python
with open('../DATA/Ames_Housing_Feature_Description.txt','r') as f:
    print(f.read())
```

```python
df=pd.read_csv('../DATA/Ames_outliers_removed.csv')
```

```python
df.info()
```

```python
df = df.drop('PID',axis=1)
```

```python
df.isnull().sum()>0
```

```python
def percent_missing(df):
    percent_nan=100*df.isnull().sum()/len(df)
    percent_nan=percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

```python
percent_nan=percent_missing(df)
```

```python
plt.figure(dpi=150)
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.xticks(rotation=90);
```

```python
percent_nan[percent_nan<1]
```

```python
df[df['Electrical'].isnull()]['Garage Area']
```

```python
df[df['Bsmt Half Bath'].isnull()]
```

```python
df=df.dropna(axis=0,subset=['Electrical','Garage Cars'])
```

```python
df
```

```python
percent_nan=percent_missing(df)
```

```python
percent_nan
```

```python
percent_nan[percent_nan<1]
```

```python
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.ylim(0,1)
plt.xticks(rotation=90);
```

```python
df[df['Bsmt Full Bath'].isnull()]
```

```python
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols]=df[bsmt_num_cols].fillna(0)

bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols]=df[bsmt_num_cols].fillna('None')
```

```python
df['Mas Vnr Type']=df['Mas Vnr Type'].fillna('None')
```

```python
df['Mas Vnr Area']=df['Mas Vnr Area'].fillna(0)
```

```python
sub=['Bsmt Qual','Bsmt Cond','BsmtFin Type 1','BsmtFin Type 2','Bsmt Exposure']
df=df.dropna(axis=0,subset=sub)
```

```python
percent_nan=percent_missing(df)
```

```python
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.title('Percent of NaN values in each column')

plt.xticks(rotation=90);
```

```python
sub2=['Garage Type','Garage Yr Blt','Garage Finish','Garage Qual','Garage Cond']
df=df.dropna(subset=sub2)
```

```python
percent_nan=percent_missing(df)
```
