```python
def quality_bill(bill,tip):
    if tip/bill>0.25:
        return 'Generous'
    else:
        return 'other'
```

```python
quality_bill(16.99,1.01)
```

```python
df1['quality']=df1[['total_bill','tip']].apply(lambda df1:quality_bill(df1['total_bill'],df1['tip']),axis=1)
df1.describe()
```

```python
df1["quality"]=np.vectorize(quality_bill)(df1["total_bill"],df1["tip"])
```
