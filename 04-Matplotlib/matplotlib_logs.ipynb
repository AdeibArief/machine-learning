```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,4)
y=2*x
```

```python
fig=plt.figure(figsize=(10,3),dpi=100)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b,linestyle='--',color=(0,0,0),marker='o',markersize=20,markerfacecolor='blue',markeredgecolor='red',markeredgewidth=4)
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_title('something',color='green');
```

```python
fig=plt.figure(dpi=100)
axes1=fig.add_axes([0,0,1,1])
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_title('something')


axes1.plot(a,b)
axes2=fig.add_axes([0.2,0.2,0.4,0.4])
axes2.set_xlabel('x axis')
axes2.set_ylabel('y axis')
axes2.set_xlim(0,50)
axes2.set_ylim(0,5)
axes2.set_title('something 2')
axes2.plot(x,y)
```

```python
fig,axes=plt.subplots()
```

```python
fig,axes=plt.subplots(ncols=2,nrows=2)
axes[0][1].plot(x,y,color='blue',linestyle='--')
axes[1][0].plot(a,b,color=(0,0,0),linestyle=':')
axes[1][1].plot(x,y)
axes[0][0].plot(a,b)
axes[0][1].set_xlabel('0,1 label')
axes[1][0].set_xlabel('1,0 label')
axes[0][0].set_xlabel('0,0 label')
axes[1][1].set_xlabel('1,1 label')

axes[0][1].set_ylabel('0,1 label')
axes[1][0].set_ylabel('1,0 label')
axes[0][0].set_ylabel('0,0 label')
axes[1][1].set_ylabel('1,1 label')
# plt.tight_layout()
fig.suptitle('total figure',fontsize='large')
fig.subplots_adjust(wspace=1,hspace=0.9)
```

# Advanced commands

```python
fig,axes=plt.subplots(1,2,figsize=(10,8))
axes[0].plot(x,x**2)
axes[0].set_title('normal')
axes[1].set_yscale("log")
axes[1].plot(np.exp(x),x)
axes[1].set_title('log');
```

```python
fig,ax=plt.subplots(figsize=(10,4))
ax.plot(x,x**2,x,x**3)
ax.set_xticks([1,2,3,4])
ax.set_yticks([0,50,100,150])
# ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)
# ax.set_xticklabels(['$\alpha$, '$\beta$', '$\gamma$', '$\delta$', '$\epsilon$'])
# ax.set_xticklabels(['zero','fifty','hundred','one fifty'])
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.plot()
ax1=ax.twinx()
ax.plot(x,x**2,lw=2,color='blue')
ax.set_ylabel('area m^2',fontsize='large',color='blue')
for lab in ax.get_yticklabels():
    lab.set_color('blue')
ax1.plot(x,x**3,lw=2,color='red',ls='--')
ax1.set_ylabel('volume m^3',fontsize='large',color='red')
for lab in ax1.get_yticklabels():
    lab.set_color('red')
```
