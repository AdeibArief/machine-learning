```python
X=df.drop('class',axis=1)
X=pd.get_dummies(X,drop_first=True)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
help(train_test_split)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model=AdaBoostClassifier(n_estimators=1)
```

```python
model.fit(X_train,y_train)
```

```python
tests=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(classification_report(tests,y_test))
```

```python
len(X.columns)
```

```python
error_rates=[]

for n in range(1,96):
    model=AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    err=1-accuracy_score(y_test,preds)
    error_rates.append(err)
```

```python
error_rates
```

```python
# sns.countplot(range(1,96),error_rates)
plt.plot(range(1,96),error_rates)
```

```python
model
```

```python
model.feature_importances_
```

```python
feat=pd.DataFrame(data=model.feature_importances_,index=X.columns,columns=['importance'])
```

```python
feat
```

```python
# sns.barplot(data=feat,x='importance')
imp_feat=feat[feat['importance']>0]
```

```python
imp_feat=imp_feat.sort_values('importance')
```

```python
plt.figure(figsize=(8,8))
sns.barplot(data=imp_feat,y=imp_feat.index,x=imp_feat['importance'])
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
help(GradientBoostingClassifier)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model=GradientBoostingClassifier()
```

```python
param_grid={
    'n_estimators':[1,10,20,40,50,100],
    'max_depth' : [3, 7]

}
```

```python
grid=GridSearchCV(model,param_grid)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
preds=grid.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
feat=grid.best_estimator_.feature_importances_
```
